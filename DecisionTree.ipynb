```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()               #Instantiate tree class
```

```python
dtc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = dtc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
import numpy as np
from sklearn import tree
import io
import pydotplus
from IPython.display import Image

dot_iris = io.StringIO()
tree.export_graphviz(dtc, out_file = dot_iris, feature_names = iris.feature_names)
graph = graph = pydotplus.graphviz.graph_from_dot_data(dot_iris.getvalue())

Image(graph.create_png())
```
